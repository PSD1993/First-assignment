```python
#.Stacked Bar Plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movies_name=['DANGAL','BAHUBALI','KGF','PUSHPA','RRR','JAWAN','PATHAN','KALKI']
budget=[70,250,100,500,336,300,250,600]
collection=[2050,1700,1150,1800,1700,1147,1045,1045]
font1={'color':'c','size':35}
font2={'color':'m','size':25}
plt.figure(figsize=(10,8))
plt.title('Higest Grossing Indiain Movies',font1)
plt.xlabel('Movies',fontdict=font2)
plt.ylabel('Budget and Collection',font2)
plt.bar(movies_name,budget,color='blue',label="BUDGET")
plt.bar(movies_name,collection,bottom=budget,color='orange',label='COLLECTION')
plt.legend()
plt.xticks(movies_name,rotation=20)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

movies_name = ['DANGAL', 'BAHUBALI', 'KGF', 'PUSHPA', 'RRR', 'JAWAN', 'PATHAN', 'KALKI']
budget = [70, 250, 100, 500, 336, 300, 250, 600]
collection = [2050, 1700, 1150, 1800, 1700, 1147, 1045, 1045]
width = 0.25
font1 = {'color': 'c', 'size': 35}
font2 = {'color': 'm', 'size': 25}


plt.title('Highest Grossing Indian Movies', font1)
plt.xlabel('Movies',font2)
plt.ylabel('Budget and Collection',font2)

# Plot budget and collection as stacked bars
plt.bar(movies_name, budget, color='blue', label="BUDGET")
plt.bar(movies_name, collection, bottom=budget, color='pink', label='COLLECTION')

plt.legend()
plt.xticks(rotation=20)
plt.show()
```

```python
movies_name = ['DANGAL', 'BAHUBALI', 'KGF', 'PUSHPA', 'RRR', 'JAWAN', 'PATHAN', 'KALKI']
budget = [70, 250, 100, 500, 336, 300, 250, 600]
collection = [2050, 1700, 1150, 1800, 1700, 1147, 1045, 1045]
width = 0.35
BG=np.arange(len(movies_name))
MC=[j+width for j in BG]
plt.title('Highest Grossing Indian Movies', color= 'purple',size=20)
plt.xlabel('Movies',color='c',size=15)
plt.ylabel('Budget and Collection',color='c',size=15)
plt.bar(BG,budget,width,color='blue', label="BUDGET")
plt.bar(MC,collection,width,color='orange', label='COLLECTION')
plt.legend()
plt.xticks(BG+width/2,movies_name,rotation=20,alpha=2
           .0)
plt.show()
```
