```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=["PYTHON","DATA SCIENCE","JAVA","WEB DEVLOPMENT"]
y=[80,98,70,80]
c=["r","y","g","m"]
plt.xlabel('Languages',fontsize=20)
plt.ylabel('fees',fontsize=20)
plt.title("TOPS TECHNOLOGY",fontsize=20)
plt.bar(x,y,width=0.4,color=c,label='Courses Offerse')
plt.legend(loc="best")
plt.show()
```

```python
#multiple bar plot

x=['PYTHON','WEB DEVLOPMENT','DATA SCIENCE','JAVA','C++']
y=[80,70,90,60,50]
z=[50,60,80,70,60]
width=0.2
p=np.arange(len(x))
p1=[j+ width for j in p]
font1={"family":'sans-serif','color':'g','size':20}
font2={"family":'sans-serif','color':'r','size':20}
font3={"family":'monospace','color':'y','size':30}
plt.xlabel("languages",font1)
plt.ylabel("No",font2)
plt.title("Tops Technology",font3)
plt.bar(p,y,width,color="g",label="Most Popular languages in 2024-25")
plt.bar(p1,z,width,color="r",label="Most Popular languages in 2010-11")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
